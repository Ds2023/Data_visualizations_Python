```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('county-outpatient-morbidity-for-patients-below-5-years-of-age-in-2013-csv-2 (1).csv')
df.head()
```

County Metrics Analysis

```python
df.info()
```

Ranking the Counties on First First Attendances, Referral Ins and Referral Outs.

```python
new = df.iloc[:,:2].sort_values(by=['First Attendances']).head()
new
```

```python
high = df.iloc[:,:2].sort_values(by=['First Attendances']).tail()
high
```

```python
ref_in = df[['County','Referrals In']].sort_values(by=['Referrals In']).head()
ref_in
```

```python
ref = df[['County','Referrals In']].sort_values(by=['Referrals In']).tail()
ref
```

```python
out = df[['County','Referrals Out']].sort_values(by=['Referrals Out']).head()
out
```

```python
out_low = df[['County','Referrals Out']].sort_values(by=['Referrals Out']).tail()
out_low
```

```python
plt.figure(figsize=(12,12))

plt.subplot(3,2,1)
plt.bar(new['County'],new['First Attendances'],color='magenta')
plt.title('Counties with the Least First Attendance Rates')
plt.ylabel('First Attendances per County')

plt.subplot(3,2,2)
plt.bar(high['County'],high['First Attendances'],color='magenta')
plt.title('Counties with the Highest First Attendance Rates')
plt.ylabel('First Attendances per County')

plt.subplot(3,2,3)
plt.bar(ref_in['County'],ref_in['Referrals In'],color='teal')
plt.title('Counties with the Least Referral In Rates')
plt.ylabel('Referrals In per County')

plt.subplot(3,2,4)
plt.bar(ref['County'],ref['Referrals In'],color='teal')
plt.title('Counties with the Most Referral In Rates')
plt.ylabel('Referrals In per County')

plt.subplot(3,2,5)
plt.bar(out['County'],out['Referrals Out'],color='gold')
plt.title('Counties with the Least Referrals Out Rates')
plt.ylabel('Referrals Out per County')

plt.subplot(3,2,6)
plt.bar(out_low['County'],out_low['Referrals Out'],color='gold')
plt.title('Counties with the Most Referrals Out Rates')
plt.ylabel('Referrals Out per County')
```
